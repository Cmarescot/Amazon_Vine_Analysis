```python
# Import Pandas
import pandas as pd 
```

```python
# Create datatframe from imported csv
vine_table_df = pd.read_csv("../../vine_table.csv")
vine_table_df
```

```python
# create a new DataFrame or table to retrieve all the rows where the total_votes count is equal to or greater than 20
helpful_reviews = vine_table_df.loc[vine_table_df["total_votes"] >= 20]
helpful_reviews
```

```python
# create a new DataFrame or table to retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
helpful_reviews_2 = helpful_reviews.loc[(helpful_reviews["helpful_votes"]/ helpful_reviews["total_votes"])>= .50]
helpful_reviews_2
```

```python
# create a new DataFrame or table that retrieves all the rows where a review was written as part of the Vine program (paid)
paid_reviews = helpful_reviews_2.loc[helpful_reviews_2["vine"] == "Y"]
paid_reviews
```

```python
# Total number of paid reviews 
number_paid = paid_reviews.count()
number_paid
```

```python
# Number of 5 star reviews (paid)
paid_5_stars = paid_reviews.loc[paid_reviews["star_rating"] == 5.0].count()
paid_5_stars
```

```python
# Percentage of 5 star reviews(paid)
# Percentage of 5 star reviews(unpaid)
percentage_5_star_paid = (paid_5_stars/number_paid)* 100
percentage_5_star_paid
```

```python
# retrieve all the rows where the review was not part of the Vine program (unpaid)
unpaid_reviews = helpful_reviews_2.loc[helpful_reviews_2["vine"] == "N"]
unpaid_reviews
```

```python
# Total number of unpaid reviews 
number_unpaid = unpaid_reviews.count()
number_unpaid
```

```python
# Number of 5 star reviews (unpaid)
unpaid_5_stars = unpaid_reviews.loc[unpaid_reviews["star_rating"] == 5.0].count()
unpaid_5_stars
```

```python
# Percentage of 5 star reviews(unpaid)
percentage_5_star = (unpaid_5_stars/number_unpaid)* 100
percentage_5_star
```
